```python
import numpy as np
import pandas as pd
```

Perform ETL

```python
hpi_df = pd.read_csv("hpi.csv")
```

```python
hpi_df['DATE']= pd.to_datetime(hpi_df['DATE'], format= '%d-%m-%Y').dt.strftime("%Y-%m-%d")
```

```python
mask = hpi_df['DATE'] <= "2023-01-01"
hpi_df = hpi_df[mask]
```

```python
hpi_df['year'] = pd.DatetimeIndex(hpi_df['DATE']).year
hpi_df['month'] = pd.DatetimeIndex(hpi_df['DATE']).month
hpi_df = hpi_df.rename(columns={"CSUSHPISA": "hpi"})
print("Shape of the S&P/Case-Shiller Index:", hpi_df.shape)
hpi_df.head()
hpi_df.tail()
```

```python
df_hu = pd.read_csv("completed_housing_unit.csv")
df_hu['DATE']= pd.to_datetime(df_hu['DATE'], format= '%d-%m-%Y').dt.strftime("%Y-%m-%d")
print("Shape of completed Housing Unit Index :", df_hu.shape)
df_hu.tail()
```

```python
df_cpi = pd.read_csv("cpi.csv")
df_cpi = df_cpi.rename(columns={"CPIAUCSL": "cpi"})
df_cpi['DATE']= pd.to_datetime(df_cpi['DATE'], format= '%d-%m-%Y').dt.strftime("%Y-%m-%d")
print("Shape of Consumer Price Index :", df_hu.shape)
df_cpi.tail()
```

```python
df_gdp = pd.read_csv("gdp.csv")
df_gdp['DATE']= pd.to_datetime(df_gdp['DATE'], format= '%d-%m-%Y').dt.strftime("%Y-%m-%d")
print("Shape of per capita GDP :", df_gdp.shape)
df_gdp.tail()
#data is quaterly. Will convert into monthly whilst creating final dataframe
```

```python
df_subsidy = pd.read_csv("govt_subsidies.csv")
df_subsidy['DATE']= pd.to_datetime(df_subsidy['DATE'], format= '%d-%m-%Y').dt.strftime("%Y-%m-%d")
df_subsidy = df_subsidy.rename(columns={"L312051A027NBEA": "subsidies"})
print("Shape of subsidies data :", df_subsidy.shape)
df_subsidy.tail()
#data is annually. Will convert into monthly whilst creating final dataframe
```

```python
df_income = pd.read_csv("income.csv")
df_income['DATE']= pd.to_datetime(df_income['DATE'], format= '%d-%m-%Y').dt.strftime("%Y-%m-%d")
df_income = df_income.rename(columns={"DSPIC96": "income"})
print("Shape of Income data :", df_income.shape)
df_income.tail()
```

```python
df_fed_ir = pd.read_csv("interest_rates.csv")
df_fed_ir['DATE']= pd.to_datetime(df_fed_ir['DATE'], format= '%d-%m-%Y').dt.strftime("%Y-%m-%d")
df_fed_ir = df_fed_ir.rename(columns={"FEDFUNDS": "fed_interest_rate"})
print("Shape of Interest rate data :", df_fed_ir.shape)
df_fed_ir.tail()
```

```python
df_med_income = pd.read_csv("median_house_income.csv")
df_med_income['DATE']= pd.to_datetime(df_med_income['DATE'], format= '%d-%m-%Y').dt.strftime("%Y-%m-%d")
df_med_income = df_med_income.rename(columns={"MEHOINUSA672N": "median_house_income"})
print("Shape of Median House Income data :", df_med_income.shape)
df_med_income.tail()
#data is annually. Will convert into monthly whilst creating final dataframe
```

```python
df_new_house = pd.read_csv("monthly_supply_of_new_house.csv")
df_new_house['DATE']= pd.to_datetime(df_new_house['DATE'], format= '%d-%m-%Y').dt.strftime("%Y-%m-%d")
print("Shape of Monthly supply of new house data :", df_new_house.shape)
df_new_house.tail()
```

```python
df_avg_mortgage = pd.read_csv("avg_mortgage.csv")
print("Shape of average mortgage value data :", df_avg_mortgage.shape)
df_avg_mortgage = df_avg_mortgage.rename(columns={"MORTGAGE30US": "avg_mortgage"})
df_avg_mortgage.tail()
```

```python
df_oldpop = pd.read_csv("old_age_pop.csv")
df_oldpop['DATE']= pd.to_datetime(df_oldpop['DATE'], format= '%d-%m-%Y').dt.strftime("%Y-%m-%d")
print("Shape of Population data above age 60 :", df_oldpop.shape)
df_oldpop.tail()
#data is annually. Will convert into monthly whilst creating final dataframe
```

```python
df_pop = pd.read_csv("population.csv")
df_pop['DATE']= pd.to_datetime(df_pop['DATE'], format= '%d-%m-%Y').dt.strftime("%Y-%m-%d")
df_pop = df_pop.rename(columns={"POPTHM": "population"})
print("Shape of Population data :", df_pop.shape)
df_pop.tail()
```

```python
df_ppic = pd.read_csv("ppic.csv")
df_ppic['DATE']= pd.to_datetime(df_ppic['DATE'], format= '%d-%m-%Y').dt.strftime("%Y-%m-%d")
print("Shape of Producer Price Index of Construction data :", df_ppic.shape)
df_ppic.tail()
```

```python
df_th = pd.read_csv("total_households.csv")
df_th['DATE']= pd.to_datetime(df_th['DATE'], format= '%d-%m-%Y').dt.strftime("%Y-%m-%d")
df_th = df_th.rename(columns={"TTLHH": "total_households"})
print("Shape of Total Households data :", df_th.shape)
df_th.tail()
#data is annually. Will convert into monthly whilst creating final dataframe
```

```python
df_cons = pd.read_csv("ttl_construction.csv")
df_cons['DATE']= pd.to_datetime(df_cons['DATE'], format= '%d-%m-%Y').dt.strftime("%Y-%m-%d")
df_cons = df_cons.rename(columns={"TTLCONS": "total_house_cons"})
print("Shape of Total Construction of Households data :", df_cons.shape)
df_cons.tail()
```

```python
df_unemp = pd.read_csv("unemployment_rate.csv")
df_unemp['DATE']= pd.to_datetime(df_unemp['DATE'], format= '%d-%m-%Y').dt.strftime("%Y-%m-%d")
df_unemp = df_unemp.rename(columns={"UNRATE": "unemp_rate"})
print("Shape of Unemployment Rate :", df_unemp.shape)
df_unemp.tail()
```

```python
hpi_df= pd.merge(hpi_df, df_gdp, how="left")
hpi_df.head()
```

```python
#merging dataframes having monthly data together
df = pd.DataFrame()
df_months=[hpi_df, df_unemp, df_hu, df_cpi, df_cons, df_ppic, df_pop, df_new_house, df_income, df_fed_ir, df_avg_mortgage]
for df1 in df_months:
  df1['DATE']= pd.to_datetime(df1['DATE'])
  df1 = df1.set_index('DATE')
  df = pd.concat([df,df1], axis = 1)
print(df.shape)
df.head()
```

```python
df_annual = [df_th, df_med_income, df_subsidy, df_oldpop]
for df1 in df_annual:
    if "year" not in df1.columns:
        df1["year"] = pd.DatetimeIndex(df1["DATE"]).year
        df1.set_index("DATE", inplace = True)
        df = pd.merge(df, df1, how = "left", on = "year")
    else:
      df = pd.merge(df, df1, how = "left", on = "year")
df["DATE"] = hpi_df["DATE"]
df.set_index("DATE", inplace = True)
df.head()
```
